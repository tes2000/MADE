ANALYSIS - CRIME DATA & WEATHER DATA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium.plugins import HeatMap
```

Data Initialization

```python
# File paths for cleaned datasets
crime_data_path = "C:/Users/HP/Documents/MADE/data/cleaned_crime_data.csv"
weather_data_path = "C:/Users/HP/Documents/MADE/data/cleaned_weather_data.csv"
```

```python
# Load Cleaned Data
crime_data = pd.read_csv(crime_data_path)
weather_data = pd.read_csv(weather_data_path)
```

```python
# Convert date columns to datetime format
crime_data['Date'] = pd.to_datetime(crime_data['Date'], errors='coerce')
weather_data['Date'] = pd.to_datetime(weather_data['Date'], errors='coerce')
```

Temporal Patterns

```python
# Daily Crime Count
daily_crime_count = crime_data.groupby(crime_data['Date'].dt.date).size()
plt.figure(figsize=(12, 6))
plt.plot(daily_crime_count, label="Daily Crime Count")
plt.title("Daily Crime Count Over Time")
plt.xlabel("Date")
plt.ylabel("Crime Count")
plt.legend()
plt.grid()
plt.show()
```

Crime Type Distribution

```python
top_crime_types = crime_data['Crime Type'].value_counts().head(10)
plt.figure(figsize=(12, 6))
top_crime_types.plot(kind='bar', color='skyblue')
plt.title("Top 10 Crime Types")
plt.xlabel("Crime Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

Spatial Analysis

```python
# Clean latitude and longitude columns
crime_data['Latitude'] = pd.to_numeric(crime_data['Latitude'], errors='coerce')
crime_data['Longitude'] = pd.to_numeric(crime_data['Longitude'], errors='coerce')
clean_crime_data = crime_data.dropna(subset=['Latitude', 'Longitude'])
```

```python
# Create a Folium heatmap
lat_lng_pairs = list(zip(clean_crime_data['Latitude'], clean_crime_data['Longitude']))
heatmap_map = folium.Map(location=[clean_crime_data['Latitude'].mean(), clean_crime_data['Longitude'].mean()], zoom_start=10)
HeatMap(lat_lng_pairs, radius=15, blur=10).add_to(heatmap_map)
```

```python
# Save heatmap to file
heatmap_map.save("crime_heatmap.html")
print("Crime heatmap saved as 'crime_heatmap.html'.")
```

Weather Analysis

```python
plt.figure(figsize=(12, 6))
plt.plot(weather_data['Date'], weather_data['tmax'], label='Max Temperature', color='red')
plt.plot(weather_data['Date'], weather_data['tmin'], label='Min Temperature', color='blue')
plt.title("Daily Max and Min Temperature Over Time")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.grid()
plt.show()
```

Correlation Analysis

```python
# Merge datasets on Date
merged_data = pd.merge(crime_data, weather_data, on='Date', how='inner')
```

```python
# Calculate correlation between weather and crime
correlation = merged_data[['tmax', 'tmin', 'Precipitation', 'Latitude', 'Longitude']].corr()
correlation
```

```python
# Plot correlation heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Weather and Crime")
plt.show()
```
